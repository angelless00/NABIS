```python
## 두 연도 비교하는 함수
def compare(DF,start,end):
    """column명 string사용"""
    gap=DF[end]-DF[start]
    M=gap.max()
    m=gap.min()
    gap=DF[end]-DF[start]
    m_dict={'max':[],'min':[]}
    for i in gap.index:
        if gap[i]==gap.max():
            m_dict['max'].append(i)
        elif gap[i]==gap.min():
            m_dict['min'].append(i)
    return m_dict
```

```python
compare(trafficDF,'18','23')
```

```python
compare(fireDF,'18','23')
```

```python
compare(criminalDF,'18','23')
```

```python
def min_max(df,year,mm):
    """mm== 'min' or 'max """
    if mm=='min':
        for i in df[year].index:
            df.loc[i,year]==df[year].min()
            print(i)
    elif mm=='max':
        for i in df[year].index:
            df.loc[i,year]==df[year].max()
            print(i)
```

```python
for idx in velocity.index:
    for ii in region.index:
        if velocity.loc[idx,'emd_code']==region.loc[ii,'읍면동코드.2']:
            velocity.loc[idx,'시도']=region.loc[ii,'시도명칭.2']
            velocity.loc[idx,'시군구']=region.loc[ii,'시군구명칭.2']
            velocity.loc[idx,'읍면동']=region.loc[ii,'읍면동명칭.2']
            break
```

```python
cc=pd.read_csv('혼잡시평균속도.csv')
cc.tail()
```

```python
cc_group=cc.groupby(['시도'])
```

```python
avgvelo=pd.DataFrame(columns=['지역','속도'])
n=0
for i in list(cc_group.groups.keys()):
    avgvelo.loc[n,'지역']=i
    avgvelo.loc[n,'속도']=cc.iloc[cc_group.indices[i]]['velocity_CG_AVRG'].mean()
    n=n+1
```

```python
avgvelo.to_csv('지역별혼잡시속도.csv',index=False,encoding='utf-8')
```
